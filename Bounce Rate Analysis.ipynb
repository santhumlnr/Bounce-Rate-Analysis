```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample session data (replace with actual data)
data = {
    "session_id": [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "user_id": [101, 102, 103, 104, 105, 106, 107, 108, 109],
    "pages_visited": [1, 3, 1, 2, 1, 4, 2, 1, 5],
    "session_duration": [30, 120, 20, 60, 15, 300, 90, 10, 400],  # Duration in seconds
}

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Calculate bounce rate
total_sessions = df.shape[0]
bounced_sessions = df[df["pages_visited"] == 1].shape[0]
bounce_rate = (bounced_sessions / total_sessions) * 100

# Visualization: Bounce vs Non-Bounce Sessions
categories = ["Bounced Sessions", "Non-Bounced Sessions"]
values = [bounced_sessions, total_sessions - bounced_sessions]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=categories, autopct="%1.1f%%", startangle=90, colors=["red", "green"])
plt.title("Bounce Rate Analysis")
plt.show()

# Print insights
print("Bounce Rate Analysis:")
print("-" * 30)
print(f"Total Sessions: {total_sessions}")
print(f"Bounced Sessions: {bounced_sessions}")
print(f"Bounce Rate: {bounce_rate:.2f}%")
```
